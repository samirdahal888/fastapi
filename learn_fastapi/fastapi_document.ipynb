# Query Paramter

- Query parameter are Optional key0value pairs appended to the end of the url, Used to pass additional data to the server in an http request

- They are typically employed for operations like filering , sorting ,searching, without altering the endpoint itself 



    - /patient?city=kathmandu & sorted_by=age

- The ? marks that starta of query parameter
- Each Parameter is a key value pair ; key=value
- Multiple Parameter are seprated by &

- in the above case city=kathmandu and sorted_by=age is a query paramter

### Query ()
