# What is API?

- Api (Application Programming interface) are mechanisms that enables two software to communicate with each others. such as front end and backend , using a defined set of rules protocols and data format

- IN SIMPLE TERMS - API is the collection of endpoints that are available in the internet




#### Why we need API ? (what problem we face before so API is introduced ) (How the work was done when there was no API)

- Before we used to write code in mololithic architecture  (means frontend and backend code in the same folder)



- Back then frontend and backend use to be the same software written in a same file 

SO LIKE this work was done before API

#### so this can be solve by API
